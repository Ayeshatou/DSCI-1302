```python
#Q1
import numpy as np #Import the numpy & assign it to np 

n = np.random.randint(1, 51, size = (100,10)) #<-- range of 1 & 51 and the size (100,10)
# Calculate All the Numbers 
m = np.mean(n) 
med = np.median(n)
high = np.max(n)
std = np.std(n)

print('Mean:', m)
print('Median:', med)
print('Maximum:',high)
print('Standard Deviation:',std)
```

```python
#Calculate Values of Each Row 
m = np.mean(n, axis = 0)
med = np.median(n, axis = 0 )
high = np.max(n, axis = 0 )
std = np.std(n, axis = 0)

print('Mean:',m)
print('Median:',med)
print('Maximum:',high)
print('Standard Deviation:',std)
```

```python
#Q2
arr = np.random.randint(low = 1, high = 51, size = (10000,)) #<---use this to make the array size & range 
#arr.resize(6)
print(arr) #<-- print the array (it all randomized )
```

```python
print('The First 10 Numbers')
print(arr[:10])

print('The Last 10 Numbers')
print(arr[-10:])

print('The Variance of Array')
var = np.var(arr)
print(var)
```

```python
print("The Difference ")
print(np.diff(arr))
```

```python
print(arr[-1] - arr[0])
```

```python
arr.sort()
res = [i for n, i in enumerate(arr) if i not in arr[:n]]

print(res)

print('First 10 Element(sorted)')
print(res[:10])

print('Last 10 Element(sorted)')
print(res[-10:])

print('The Variance (sorted) ')
print(np.var(arr))
```
